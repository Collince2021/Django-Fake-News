```python
##import libraries
import pandas as pd
import string
import scipy
from scipy.io import arff
import sklearn

##models for preprocessing, classification and training
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

##from sklearn import datasets
from sklearn import datasets 
##import dataset in openml
from sklearn.datasets import fetch_openml
##specify the name of the dataset and the version

#fakenews_dataset_sample1=fetch_openml(name="Fake-News",version=1)
##DATA EXTRACTION
#data1=fakenews_dataset_sample1['data']
dataset=pd.read_csv("fake_or_real_news.csv")


## print(fakenews_dataset_sample1)
data1=dataset
```

```python
data3=data1
data3=data3.drop(columns="title")
##data1["title"]=data1["title"].str.lower()
data3["text"]=data1["text"].str.lower()
data3["label"]=data1["label"].str.lower()

data3
```

```python
#Removal of punctuations
def removePunctuation(text):
    punctuation=string.punctuation
    return text.translate(str.maketrans("","",punctuation))

##data1["title"]=data1["title"].apply( lambda x:removePunctuation(x))
data3["text"]=data3["text"].apply(lambda x:removePunctuation(x))

##data1.fillna("Unknown", inplace=True)
data3
```

```python
##Removal of stopwords
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")

list_stopwords=set(stopwords.words("english"))
def removeStopwords(text):
    return  " ".join([word for word in text.split() if word not in list_stopwords])

##data1["title"]=data1["title"].apply( lambda x:removeStopwords(x))
data3["text"]=data3["text"].apply(lambda x:removeStopwords(x))
#data3["text"]=data1["text"].apply(lambda x:removePunctuation(x))
data3
```

```python
import re
def remove_Specailcharacters(text):
    text=re.sub('[^a-zA-Z0-9]'," ",text)
    text=re.sub("\s"," ",text)
    return text

#data3["title"]=data1["title"].apply( lambda x:remove_Specailcharacters(x))
data3["text"]=data3["text"].apply(lambda x:remove_Specailcharacters(x))
data3
```

```python
##lematization
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
#nltk.download('punkt')
#nltk.download('wordnet')
import nltk
nltk.download('punkt')


# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmatize_text(text):
    # Tokenize the text into words
    tokens = word_tokenize(text)
       # Lemmatize each word in the text
    lemmatized_text = [lemmatizer.lemmatize(word) for word in tokens]
    # Join the lemmatized words back into a single string
    lemmatized_text = ' '.join(lemmatized_text)
    return lemmatized_text

# Perform lemmatization

##data3["title"]=data1["title"].apply(lambda x:lemmatize_text(x))
data3["text"]=data3["text"].apply(lambda x:lemmatize_text(x))


data3
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer

# Split the data into features text and labels
X = data3['text']
y = data3['label']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Define the pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),  # Text vectorization(perfoms text cleaning)
    ('clf', RandomForestClassifier())  # Random Forest classifier
])

# Fit the pipeline to the training data
pipeline.fit(X_train, y_train)

# Predict labels for the testing data
y_pred = pipeline.predict(X_test)


```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
